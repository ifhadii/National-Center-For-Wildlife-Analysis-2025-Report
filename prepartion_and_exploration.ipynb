```python
import pandas as pd
import openpyxl
```

```python
df = pd.read_excel('data/wildlife_launch_and_resettlement_dataset.xlsx')
df
```

```python
# Identify null values
print(df.isnull().sum())
```

```python
file_path = "data/wildlife_launch_and_resettlement_dataset.xlsx"
with pd.ExcelWriter(file_path, engine='openpyxl', mode='w') as writer:
    df.to_excel(writer, index=False)
```
